```python

df = pd.read_csv('csvs/mean_years_in_school_men_15_to_24_years.csv', index_col='country')
df.columns = df.columns.astype(int)
df.head()
```

```python
men_df = pd.read_csv('csvs/mean_years_in_school_men_15_to_24_years.csv', index_col='country')
women_df = df = pd.read_csv('csvs/mean_years_in_school_women_15_to_24_years.csv', index_col='country')


```

```python
len(men_df)==len(women_df)
```

```python
men_df.head()
```

```python
women_df.head()
```

```python
df_concat = pd.concat((men_df, women_df))
by_row_index = df_concat.groupby(df_concat.index)
df_means = by_row_index.mean()

df_means.head()
```

```python
len(df_means)
```

```python
new_data=[]
df = df_means
# df = pd.read_csv('csvs/mean_years_in_school_men_15_to_24_years.csv', index_col='country')
df.columns = df.columns.astype(int)

minv,maxv=999999,-999999

for i in range(len(data)):
    country_name = data[i]['name']
    if country_name in df.index:
        new_data.append(data[i])
        curr_dict = dict(df.loc[country_name])
        curr_list = []
        for k,v in curr_dict.items():
            minv = min(minv,v)
            maxv=max(maxv,v)
            curr_list.append([k,v])
        new_data[-1]['mean_years_in_school_15_to_24_years'] = curr_list
```

```python
len(new_data),minv,maxv
```

```python
with open('new_nations.json', 'w') as fout:
    json.dump(new_data, fout)
```
