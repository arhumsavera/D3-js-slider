```python
#df = df_means
new_data=[]
df =df.fillna(0)
df.columns = df.columns.astype(int)

minv,maxv=999999,-999999

for i in range(len(data)):
    country_name = data[i]['name']
    if country_name in df.index and country_name !='Brunei' :
        new_data.append(data[i])
        curr_dict = dict(df.loc[country_name])
        curr_list = []
        for k,v in curr_dict.items():
            minv = min(minv,v)
            maxv=max(maxv,v)
            curr_list.append([k,v])
        new_data[-1]["co2_emissions_tonnes_per_person"] = curr_list
```
